```python
fig = plt.figure(figsize=(12, 6))
sqft = fig.add_subplot(121)
cost = fig.add_subplot(122)

sqft.hist(df.sqft_living, bins=80)
sqft.set_xlabel('Feet^2')
sqft.set_title("Histogram of Living Room Per Square Feet")

cost.hist(df.price, bins=80)
cost.set_xlabel('Price($)')
cost.set_title("Histogram of Housing Prices")

plt.show()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
model = ols('price ~ sqft_living', df).fit()
print(model.summary())
```

```python
model = ols('price ~ sqft_living + bedrooms + grade + condition', df).fit()
print(model.summary())
```

```python
sns.jointplot(x = "sqft_living", y = "price", data = df, kind = 'reg', fit_reg = True, size = 8)
plt.show()
```
